```python
from sklearn.feature_extraction.text import TfidfVectorizer
import json
from tqdm import tqdm
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
with open('mehrakhbar-1400.json' , 'r') as f:
  doc = json.load(f)
```

```python
len(doc) , type(doc) , doc[1]
```

```python
sl = [d['title'] for d in doc]
```

```python
sl[3]
```

```python
# tf-idf
vectorizer = TfidfVectorizer()
tfidf_docs = vectorizer.fit_transform(sl)

```

```python
tfidf_docs.shape, len(vectorizer.vocabulary_)

```

```python
list(vectorizer.vocabulary_.keys())[:10]

```

```python
query ='واکسن کرونا'

tfidf_query = vectorizer.transform([query])[0]
```

```python
# similarities
cosines = []
for d in tqdm(tfidf_docs):
  cosines.append(float(cosine_similarity(d, tfidf_query)))

```

```python
# sorting
k = 15
sorted_ids = np.argsort(cosines)
for i in range(k):
  cur_id = sorted_ids[-i-1]
  print(sl[cur_id], cosines[cur_id])
```
